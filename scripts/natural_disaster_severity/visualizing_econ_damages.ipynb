```python
# importing relevant packages
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
nat_disas_gdp = pd.read_csv("../output/nat_disas_with_gdp.csv")
nat_disas_gdp['year'] = nat_disas_gdp['start_date'].str.extract(r"(\d+)(?=-)")
nat_disas_gdp
```

```python
# Group by year across all countries
yearly = (
    nat_disas_gdp.groupby('year')['total_affected']
    .head(10)
)

plt.figure(figsize=(8,5))
plt.plot(yearly['year'], yearly['gdp_percent'] / 1000, marker='o')
plt.title('Global Disaster Damages (2020–2025)')
plt.xlabel('Year')
plt.ylabel('Total Damages (in billions USD)')
plt.grid(True)
plt.show()
```

```python
top_countries_list = ["Colombia", "Brazil", "Haiti", "Italy", "Japan"]
top_country_df = nat_disas_gdp[nat_disas_gdp.Country.isin(top_countries_list)]
top_country_df
```

```python
percent_change_country = nat_disas_gdp.groupby("Country")["gdp_percent_change"].sum().sort_values(ascending = True)

plt.title('Climate Sentiment Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Mean Sentiment')
plt.legend()
plt.grid(True)
plt.show()
```
